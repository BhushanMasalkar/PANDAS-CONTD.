```python
#importing the libraries
import numpy as np 
import pandas as pd
```

```python
#reading csv file
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
print(df)
```

```python
#question no.:-1) Deleting unnamed columns
df.drop(columns='Unnamed: 0',axis=1,inplace=True)
print(df)
```

```python
#Question no:-2 Showing the distribution of male and female
gender_categ=pd.Categorical(df['Gender']) 
gender_categ.describe() 
```

```python
#Question no.:-3 Showing the top 5 most preferred names
df['Name'].value_counts().head()
```

```python
#Question no.:-4 Showing the median name occurence in the dataset
top=df.median().Id   
name = df[df['Id']==top].Name  
Na=list(name)[0]  
df1 = df[df['Name']==Na]   
df1['Name'].value_counts() 
```

```python
#Question no.:-5 Showing the distribution of male and female born count by states
df1=df.groupby(['State','Gender'])
df1.count()
```
